```python
import gradio as gr
```

```python
def detect_emergency(file):
    # Determine if file is image or video
    if file.name.lower().endswith((".jpg", ".jpeg", ".png")):
        results = model(file.name)
        res_plotted = results[0].plot()
        output_path = "image_with_boxes.jpg"
        cv2.imwrite(output_path, res_plotted)
        return output_path
    else:
        cap = cv2.VideoCapture(file.name)
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')
        output_path = "video_with_boxes.mp4"
        out = cv2.VideoWriter(output_path, fourcc,
                              cap.get(cv2.CAP_PROP_FPS),
                              (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)),
                               int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))))
        emergency_detected = False
        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break
            results = model.predict(frame, conf=0.5, verbose=False)
            res_plotted = results[0].plot()
            out.write(res_plotted)
            for box in results[0].boxes:
                class_name = model.names[int(box.cls)]
                if class_name.lower() in ["ambulance", "firetruck", "policecar"]:
                    emergency_detected = True
        cap.release()
        out.release()
        return output_path
```

```python
iface = gr.Interface(
    fn=detect_emergency,
    inputs=gr.File(label="Upload Image or Video"),
    outputs=gr.File(label="Processed Output"),
    title="Emergency Vehicle Detection",
    description="Upload an image or video to detect emergency vehicles (ambulance, firetruck, policecar)."
)

iface.launch()
```
